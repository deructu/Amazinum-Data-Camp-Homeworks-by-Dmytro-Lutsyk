# Task about prods and sums

```python
def search_all_pairs():
    ''' Func seacrhing all pairs from 2 to 99'''
    
    pairs = [(a, b) for a in range(2, 100) for b in range(a, 100)]
    return pairs

pairs = search_all_pairs()
```

```python
def sums_and_pruducts(pairs):
    ''' Func calculation all prods and sums'''

    dict_prods = {(a, b): a * b for a, b in pairs}
    dict_sums = {(a, b): a + b for a, b in pairs}
    return dict_prods, dict_sums

dict_prods, dict_sums = sums_and_pruducts(pairs)
    
```

```python
def alg(prods, sums):
    ''' Func implements full algorithm-searching'''

    # Searching not unique prods for pair
    not_unique_prod = []
    for a, b in pairs:
        prod = prods[(a, b)]
        counter = 0
        for x, y in pairs:
            if prods[(x, y)] == prod:
                counter += 1
        if counter >= 2:
            not_unique_prod.append((a, b))


    # Searching pairs where sum knows that product is not unique
    sum_know_it = []
    for a, b in not_unique_prod:
        sum_value = sums[(a, b)]
        flag = True

        for x, y in pairs:
            if sums[(x, y)] == sum_value and len([1 for i, j in pairs if prods[(i, j)] == prods[(x, y)]]) == 1:
                flag = False
                break
        if flag:
            sum_know_it.append((a, b))


    # Searching pairs where product becomes unique after previous step
    prod_know_it = []
    for a, b in sum_know_it:
        prod = prods[(a, b)]
        counter = 0
        for x, y in sum_know_it:
            if prods[(x, y)] == prod:
                counter += 1

        if counter == 1:
            prod_know_it.append((a, b))


    # Searching pairs where sum also becomes unique after previous step
    sum_know_it_too = []
    for a, b in prod_know_it:
        s = sums[(a, b)]
        counter = 0
        for x, y in prod_know_it:
            if sums[(x, y)] == s:
                counter += 1

        if counter == 1:
            sum_know_it_too.append((a, b))

    return sum_know_it_too
```

```python
print('Result: ', *alg(dict_prods, dict_sums))
```
