````python
import pandas as pd
import numpy as np
````

````python
def answer_one():
    '''Ця функція виконує читання 3-х датасетів, 
      препроцесинг, та повертає їх об'єднання'''

    
    energy = (pd
        .read_excel('Energy_Indicators.xls')
        .drop(['Unnamed: 0', 'Unnamed: 1'], axis=1)
        .replace('...', np.nan)
        .iloc[16:244]
        .reset_index(drop=True)
    )
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy['Energy Supply'] *= 1_000_000
    energy['Country'] = (energy['Country']
        .replace('[0-9]+', '', regex=True)
        .str.extract(r'^(.+?)(?:\s*\([^)]*\))?$')[0]
        .replace({'Republic of Korea'                                   : 'South Korea',
                  'United States of America'                            : 'United States',
                  'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
                  'China, Hong Kong Special Administrative Region'      : 'Hong Kong'
                 })
    )


    GDP = pd.read_csv('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_133326.csv',  skiprows=3).rename(columns={'Country Name': 'Country'})
    GDP['Country'] = (GDP['Country']
        .replace({'Korea, Rep.'         : 'South Korea',
                  'Iran, Islamic Rep.'  : 'Iran',
                  'Hong Kong SAR, China': 'Hong Kong'})
    )
    GDP_merge_cols = ['Country'] + [str(year) for year in range(2006, 2016)] # Тільки колонки Country та list comprehension для років від 2006 до 2015
    

    ScimEn = (pd
        .read_excel('scimagojr_country_rank_1996-2024.xlsx')
        .drop('Region', axis=1) 
    )
    ScimEn_top15_rank = ScimEn[ScimEn['Rank'] <= 15]

    
    merged_dfs = (pd
        .merge(pd.merge(ScimEn_top15_rank, energy, how='inner', on='Country'), GDP[GDP_merge_cols], how='inner', on='Country')
        .set_index('Country')
                 
    )
    
    return merged_dfs
````

### Question 2
What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

*This function should return a Series named `avgGDP` with 15 countries and their average GDP sorted in descending order.*

````python
def answer_two():
    ''' Ця функція повертає Series, середнє значення
       ВВП для всіх країн від 2006 до 2015 років '''

    
    Top15 = answer_one()
    years = [str(year) for year in range(2006, 2016)] # list comprehension для років від 2006 до 2015
    
    avgGDP = (Top15[years]
        .dropna()
        .mean(axis=1)
        .sort_values(ascending=False) 
    )
    
    return avgGDP
````

````python
def answer_three():
    ''' Ця функція повертає чило, яке показу аскільки змінився ВВП
        за 10 років для країни з 6-м за розміром середнім ВВП '''
    
    
    Top15 = (answer_one()
        .assign(mean_GDP=answer_two())
        .sort_values('mean_GDP', ascending=False)
    
    )
    
    res = Top15['2015'].iloc[5] - Top15['2006'].iloc[5]
    return res
````

### Question 4

Create a new column that is the ratio of Self-Citations to Total Citations. 
What is the maximum value for this new column, and what country has the highest ratio?

*This function should return a tuple with the name of the country and the ratio.*

````python
def answer_four():
    ''' Ця функція повертає назву країни та найбільше відношення Self-Citations / Total Citations '''

    
    Top15 = answer_one()

    Top15['Self-Citations / Total Citations'] = Top15['Self-citations'] / Top15['Citations']
    index_of_max_ratio = Top15['Self-Citations / Total Citations'].argmax()
    
    return (Top15.iloc[index_of_max_ratio].name, Top15['Self-Citations / Total Citations'].iloc[index_of_max_ratio])
````

*This function should return a single string value.*

### Question 6
Create a column that estimates the number of citable documents per person. 
What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the `.corr()` method, (Pearson's correlation).

*This function should return a single number.*


````python
def answer_six():
    Top15 = (answer_one()
        .assign(Population=lambda df_: df_['Energy Supply'] / df_['Energy Supply per Capita'])
        .assign(Docs_per_capita=lambda df_: df_['Citable documents'] / df_['Population'])
    )
    
    
    return Top15['Docs_per_capita'].corr(Top15['Energy Supply per Capita'])
````

### Question 7
Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
def answer_seven():

    ContinentDict  = {'China'             :'Asia', 
                      'United States'     :'North America', 
                      'Japan'             :'Asia', 
                      'United Kingdom'    :'Europe', 
                      'Russian Federation':'Europe', 
                      'Canada'            :'North America', 
                      'Germany'           :'Europe', 
                      'India'             :'Asia',
                      'France'            :'Europe', 
                      'South Korea'       :'Asia', 
                      'Italy'             :'Europe', 
                      'Spain'             :'Europe', 
                      'Iran'              :'Asia',
                      'Australia'         :'Australia', 
                      'Brazil'            :'South America'
    }
    
    Top15 = (answer_one()
            .assign(Population=lambda df_: df_['Energy Supply'] / df_['Energy Supply per Capita'])
    )

    Top15['Continent'] = Top15.index.map(ContinentDict)

    
    return Top15.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
````
