```python
import functools

@functools.lru_cache(maxsize=None)  # Decorator for saving results
def max_floors(glass, drops):
    if drops == 1 or glass == 1: 
        return drops
    return 1 + max_floors(glass - 1, drops - 1) + max_floors(glass, drops - 1)

def drop(floors, glass):
    left, right = 1, floors
    while left < right:
        mid = (left + right) // 2
        if max_floors(glass, mid) >= floors:
            right = mid
        else:
            left = mid + 1
    return left

print("100 / 2 =", drop(100, 2))      
print("100 / 3 =", drop(100, 3))        
print("1000 / 3 =", drop(1000, 3))    
print("1000 / 5 =", drop(1000, 5)) 
```
